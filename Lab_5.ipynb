#Given:
(n)=7f(n−1)−10f(n−2),f(0)=1,f(1)=3

Analytical

r^2=7r-10

r^2−7r+10=0

(r−2)(r−5)=0

The roots are r_1=5 and r_2=2


General Solution:

f(n)=A*5^n+B*2^n


Solve for A and B (Initial Conditions):

f(0)=1:

f(1)=5A+2B=3

A+B=1

5A+2B=3


Solving the system: A=1−B.

Substitute into the second equation:

= 5A+2(1−A)=3

= 5A+2-2A=3

= 3A=1


A=1/3

B=1-1/3

 =2/3



Analytic Equation

f(n)= 1/3 * 5^n + 2/3 * 2^n


```python
# Analytic Method
# f(n) = 7f(n-1) - 10f(n-2), with f(0)=1, f(1)=3

def f_analytic(n):
    """Analytic solution for f(n) = (1/3)*5^n + (2/3)*2^n"""
    return (1/3)*(5**n) + (2/3)*(2**n)

# Test the function
for i in range(6):
    print(f"f({i}) = {f_analytic(i)}")






```

```python
#Dynamic Programming


def f_dp(n):
    if n == 0:
        return 1
    elif n == 1:
        return 3
    else:
        return 7 * f_dp(n-1) - 10 * f_dp(n-2)

# Print sample output for n from 0 to 9
for n in range(10):
    print(f"f({n}) = {f_dp(n)}")


```

```python
## 3. Dynamic Programming with Memoization


def f_memo(n):
    if n == 0:
        return 1
    elif n == 1:
        return 3
    memo = [1, 3] + [0]*(n-1)
    for i in range(2, n+1):
        memo[i] = 7 * memo[i-1] - 10 * memo[i-2]
    return memo[n]

# Print sample output for n from 0 to 9:
for n in range(10):
    print(f"f({n}) = {f_memo(n)}")


```

```python
# 4. Compare the results and discuss
print("n | Analytic | DP (Recursive) | Memoization")
print("--|----------|----------------|---------")
for n in range(10):
    a = f_analytic(n)
    b = f_dp(n)
    c = f_memo(n)
    print(f"{n:1} | {a:8.0f} | {b:14.0f} | {c:9.0f}")

print("\nDiscussion:")
print("Analytic formula computes any n instantly, and is best for very large n (e.g., n=1000).")
print("Dynamic programming (recursive) is the slowest for large n due to repeated recalculations.")
print("Memoization is both fast and efficient even for very large n, as previous results are reused.")
```
