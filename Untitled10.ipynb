```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

print("=== Linear Programming Problem ===")
print("Maximize z = x + y")
print("Subject to:")
print("  1) x + y ≤ 4")
print("  2) 8x + y ≥ 6")
print("  3) x ≥ 0, y ≥ 0\n")

# ----------------------------------------------------
# 1. Intro of Slack Variables
# ----------------------------------------------------
print("=== Slack Variable Form ===")
print("To convert inequalities to equalities:")
print("x + y + s1 = 4        (s1 ≥ 0)   # for ≤ constraint")
print("8x + y - s2 = 6        (s2 ≥ 0)   # for ≥ constraint")
print("x, y, s1, s2 ≥ 0\n")

# ----------------------------------------------------
# 2. Finding Corner Points (by solving line intersections)
# ----------------------------------------------------
# Intersection of x + y = 4 and 8x + y = 6
# Subtract: 7x = 2 -> x = 2/7, y = 4 - 2/7 = 26/7
v1 = (2/7, 26/7)

# Intersection of x + y = 4 with y = 0  -> x = 4
v2 = (4, 0)

# Intersection of 8x + y = 6 with y = 0 -> x = 3/4
v3 = (3/4, 0)

vertices = [v1, v2, v3]
print("=== Corner Points (Feasible Vertices) ===")
for i, v in enumerate(vertices, 1):
    print(f"Vertex {i}: (x, y) = ({v[0]:.4f}, {v[1]:.4f})")
print()

# ----------------------------------------------------
# 3. Graph of the Feasible Region
# ----------------------------------------------------
x = np.linspace(0, 5, 400)
y1 = 4 - x        # from x + y = 4
y2 = 6 - 8*x      # from 8x + y = 6

plt.figure(figsize=(6,6))
plt.plot(x, y1, label=r'$x + y = 4$')
plt.plot(x, y2, label=r'$8x + y = 6$')

# Fill feasible region: y between y2 and y1, y >= 0
y_lower = np.maximum(6 - 8*x, 0)
y_upper = np.minimum(4 - x, 5)
plt.fill_between(x, y_lower, y_upper, where=(y_upper >= y_lower),
                 color='gray', alpha=0.3, label='Feasible Region')

# Plot vertices
for vx, vy in vertices:
    plt.scatter(vx, vy, color='red')
    plt.text(vx + 0.05, vy, f"({vx:.2f},{vy:.2f})")

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()
plt.grid(True)
plt.title('Feasible Region for x+y≤4, 8x+y≥6, x≥0, y≥0')
plt.show()

# ----------------------------------------------------
# 4. Evaluate Objective Function at Each Corner
# ----------------------------------------------------
print("=== Evaluate Objective Function z = x + y ===")
z_values = []
for v in vertices:
    z = v[0] + v[1]
    z_values.append(z)
    print(f"At {v}: z = {z:.4f}")

max_idx = np.argmax(z_values)
max_vertex = vertices[max_idx]
print(f"\nMaximum z = {z_values[max_idx]:.4f} occurs at {max_vertex}\n")

# ----------------------------------------------------
# 5. Solve Using Built-in LP Solver (linprog)
# ----------------------------------------------------
print("=== Solving with SciPy linprog ===")
# linprog minimizes, so we minimize -x - y to maximize x + y
c = [-1, -1]                    # coefficients for -x - y
A_ub = [[1, 1], [-8, -1]]       # inequalities: x+y ≤ 4,  -8x - y ≤ -6
b_ub = [4, -6]
bounds = [(0, None), (0, None)]

res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')

print("Optimal solution from linprog:")
print(f"  x = {res.x[0]:.4f}, y = {res.x[1]:.4f}")
print(f"  Maximum z = {-res.fun:.4f}\n")

# ----------------------------------------------------
# 6. Comparing Manual vs Solver Results
# ----------------------------------------------------
print("=== Comparison ===")
print(f"Manual corner giving max: {max_vertex}, z = {z_values[max_idx]:.4f}")
print(f"Solver corner (linprog):  ({res.x[0]:.4f}, {res.x[1]:.4f}), z = {-res.fun:.4f}")

if np.allclose(res.x, max_vertex, atol=1e-6):
    print("\n The solver result matches the manual corner point exactly.")
else:
    print("\n The solver result differs — check constraints or intersection points.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Variables: x = deluxe, y = standard

# Lines for plotting
x_vals = np.linspace(0, 40, 400)
y_hours = (828 - 24*x_vals)/12     # from 24x + 12y = 828
y_material = (3654 - 84*x_vals)/70 # from 84x + 70y = 3654

plt.figure(figsize=(7,6))
plt.plot(x_vals, y_hours, label=r'$24x + 12y = 828$ (hours)')
plt.plot(x_vals, y_material, label=r'$84x + 70y = 3654$ (material)')

# Feasible region (both constraints <= and x,y>=0)
# Compute y_upper = min(y_hours, y_material) clipped at >=0
y_upper = np.minimum(y_hours, y_material)
y_upper = np.maximum(y_upper, 0)
plt.fill_between(x_vals, 0, y_upper, where=(y_upper>=0), color='gray', alpha=0.3, label='Feasible region')

# Mark vertices
vertices = [(0, 3654/70), (21,27), (828/24, 0)]
for vx, vy in vertices:
    plt.scatter(vx, vy, color='red')
    plt.text(vx+0.5, vy+0.5, f'({vx:.2f}, {vy:.2f})')

plt.xlim(0, 40)
plt.ylim(0, 60)
plt.xlabel('x = deluxe models')
plt.ylabel('y = standard models')
plt.legend()
plt.grid(True)
plt.title('Feasible Region and Vertices')
plt.show()

# Verifying with linprog (maximize R = 342x + 190y -> minimize -R)
c = [-342, -190]  # linprog minimizes
A_ub = [[24, 12], [84, 70]]
b_ub = [828, 3654]
bounds = [(0, None), (0, None)]

res = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')
print("linprog result x,y:", res.x)
print("Max revenue:", -res.fun)
```
